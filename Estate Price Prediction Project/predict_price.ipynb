```python
##Hàm kiểm tra xem x có phải là số float hay không

def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

Ý tưởng thực thi hàm xử lý cho cột total_sqft là chúng ta sẽ khai báo 1 token mà trong đó chứa các thành phần của giá trị đang xét nếu như mà biến token đang chứa 2 giá trị có nghĩa là dữ liệu đang xét ở dạng khoảng [x-y] do đó chúng ta return về giá trị trung bình của 2 số đó. Và trả về kiểu float cho dữ liệu đang xét nếu token không phải 2 giá trị ngoài 2 trường hợp kia ra thì trả về None

```python
def preprocess_total_sqft(x):
    token = x.split('-')
    if len(token)==2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except: return None
```

```python
data['total_sqft'] = data['total_sqft'].apply(preprocess_total_sqft)
```

```python
def remove_pps_outlier(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        mean = np.mean(subdf.price_per_sqft)
        std = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(mean-std)) & (subdf.price_per_sqft<=(mean+std))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out
```

```python
def bhk_outlier(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices ,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices, axis = 'index')
```

```python
remove_outlier_data = bhk_outlier(new_data)
remove_outlier_data.shape
```

```python
remove_outlier_data = remove_outlier_data[remove_outlier_data.bath<remove_outlier_data.bhk+2]
```

```python
remove_outlier_data
```

```python
dummy_variables = pd.get_dummies(remove_outlier_data.location)
```

```python
merged = pd.concat([remove_outlier_data, dummy_variables.drop(['other'], axis = 'columns')], axis = 'columns')
```

```python
final = merged[merged.bath<merged.bhk+2].drop(['location'], axis = 'columns')
```

```python
final
```

```python
X = final.drop(['price', 'price_per_sqft'], axis = 'columns')
y = final.price
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
from sklearn.model_selection import ShuffleSplit

cv = ShuffleSplit(n_splits = 5, test_size = 0.3, random_state = 0)
```

```python
cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
def predict_price(location, sqft, bath, bhk):
    loc_index = np.where(X.columns==location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0: 
        x[loc_index] = 1
    return lr.predict([x])[0]
```
